Реализовать методы поиска в соответствии с заданием. Организовать генерацию начального набора случайных данных. Для всех вариантов добавить реализацию добавления, поиска и удаления элементов.
```python
import random
import copy
import time
```

```python
#Генерация набора случайных данных
def create_matrix():
    k = int(input('Введите кол-во элементов массива: '))
    matrix = [random.randint(0, 1000) for i in range(k)]
    return matrix

matrix = create_matrix()
```

```python
#Бинарный поиск
start_time = time.time()

def binary_search(mas, n):
    low = 0;
    high = len(mas)-1
    while low <= high:
        mid = (low+high)//2
        if mas[mid] == n:
            return mid
        elif n < mas[mid]:
            high = mid-1
        else: 
            low = mid+1
    return -1

mas = copy.deepcopy(matrix)
mas = sorted(mas)
print(mas)
n = int(input('Введите элемент, который нужно найти: '))

result = binary_search(mas, n)
if result == -1:
    print('Такого элемента нет в массиве!', '\n')
else: 
    print('Индекс искомого элемента: ', result, '\n')
    
print("--- %s ms ---" % ((time.time() - start_time)*1000))
    
new = int(input('Введите элемент, который хотите добавить в массив: '))
place_new = int(input('Введите индекс элемента, который хотите добавить в массив: '))
mas.insert(place_new, new)
print(mas, '\n')

place_delete = int(input('Введите индекс элемента, который хотите удалить из массива: '))
mas.pop(place_delete)
print(mas, '\n')
```

```python
#Бинарное дерево
start_time = time.time()

class Node:

    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

    def search(self, lkpval):
        if lkpval < self.data:
            if self.left is None:
                return "Такого элемента нет в массиве!"
            return self.left.search(lkpval)
        elif lkpval > self.data:
            if self.right is None:
                return "Такого элемента нет в массиве!"
            return self.right.search(lkpval)
        else:
            return "Элемент присутсвует в массиве" 

    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print(self.data),
        if self.right:
            self.right.PrintTree()
            
mas = copy.deepcopy(matrix)         
root = Node(mas[0])
for i in range (1, len(mas)-1):
    root.insert(mas[i])
root.PrintTree() 

n = int(input('Введите элемент, который нужно найти: '))
print(root.search(n))

print("--- %s ms ---" % ((time.time() - start_time)*1000))
```

```python
#Метод Фибоначчи
start_time = time.time()

def fibonacci_search(mas, n):
    fibM_minus2 = 0
    fibM_minus1 = 1
    fibM = fibM_minus2 + fibM_minus1
    while fibM < len(mas):
        fibM_minus2 = fibM_minus1
        fibM_minus1 = fibM
        fibM = fibM_minus1 + fibM_minus2
    index = -1
    while fibM > 1:
        i = min(index + fibM_minus2, len(mas)-1)
        if mas[i] < n:
            fibM = fibM_minus1
            fibM_minus1 = fibM_minus2
            fibM_minus2 = fibM - fibM_minus1
            index = i
        elif mas[i] > n:
            fibM = fibM_minus2
            fibM_minus1 = fibM_minus1 - fibM_minus2
            fibM_minus2 = fibM - fibM_minus1
        else :
            return i
    if(fibM_minus1 and index < (len(mas)-1) and mas[index+1] == n):
        return index+1;
    return -1

mas = copy.deepcopy(matrix)
mas = sorted(mas)
print(mas)
n = int(input('Введите элемент, который нужно найти: '))

result = fibonacci_search(mas, n)
if result == -1:
    print('Такого элемента нет в массиве!', '\n')
else: 
    print('Индекс искомого элемента: ', result, '\n')
    
print("--- %s ms ---" % ((time.time() - start_time)*1000))
    
new = int(input('Введите элемент, который хотите добавить в массив: '))
place_new = int(input('Введите индекс элемента, который хотите добавить в массив: '))
mas.insert(place_new, new)
print(mas, '\n')

place_delete = int(input('Введите индекс элемента, который хотите удалить из массива: '))
mas.pop(place_delete)
print(mas, '\n')
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
board = [[0 for i in range(8)] for j in range(8)]

#Добавление ферзя в клетку [i, j]
def setQueen(i, j):
    for x in range(8):
        board[x][j] += 1
        board[i][x] += 1
        if 0 <= i + j - x < 8:
            board[i + j - x][x] += 1
        if 0 <= i - j + x < 8:
            board[i - j + x][x] += 1
    board[i][j] = -1

#Удаление ферзя
def removeQueen(i, j):
    for x in range(8):
        board[x][j] -= 1
        board[i][x] -= 1
        if 0 <= i + j - x < 8:
            board[i + j - x][x] -= 1
        if 0 <= i - j + x < 8:
            board[i - j + x][x] -= 1
    board[i][j] = 0
    
#Вывод решения
def printPosition():
    abc = 'abcdefgh'
    ans = []
    for i in range(8):
        for j in range(8):
            if board[i][j] == -1:
                ans.append(abc[j] + str(i + 1))
    print(', '.join(ans))
    
def solve(i):
    for j in range(8):
        if board[i][j] == 0:
            setQueen(i, j)
            if i == 7:
                printPosition()
            else:
                solve(i+ 1) 
            removeQueen(i, j)
                
solve(0)
```
